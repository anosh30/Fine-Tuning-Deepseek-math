```python
import pandas as pd
from datasets import Dataset

# Load the dataset
df = pd.read_csv("math_riddles.csv")

# Combine question and answer into a single text format
df["text"] = "Q: " + df["question"] + " A: " + df["answer"]

# Convert to Hugging Face Dataset format
dataset = Dataset.from_pandas(df[["text"]])

# Split into train and validation sets
dataset = dataset.train_test_split(test_size=0.2)

print(dataset)
```

```python
from transformers import AutoTokenizer

# Load the GPT-2 tokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# Fix: Set padding token to EOS token
tokenizer.pad_token = tokenizer.eos_token

# Tokenization function
def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        padding="max_length",  # Ensures consistent input size
        truncation=True,
        max_length=128
    )

# Apply tokenization
tokenized_datasets = dataset.map(tokenize_function, batched=True)

# Remove unprocessed text column
tokenized_datasets = tokenized_datasets.remove_columns(["text"])
tokenized_datasets.set_format("torch")
```

```python
import os
from transformers import AutoModelForCausalLM, TrainingArguments, Trainer

# Disable Weights & Biases (W&B) logging
os.environ["WANDB_DISABLED"] = "true"

# Load GPT-2 model
model = AutoModelForCausalLM.from_pretrained("gpt2")

training_args = TrainingArguments(
    output_dir="./math-riddle-model",
    eval_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    report_to="none"  # Fix for W&B issue
)


# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
)

# Start Training
trainer.train()
```

```python
def tokenize_function(examples):
    tokens = tokenizer(
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=128
    )
    tokens["labels"] = tokens["input_ids"].copy()  # Fix: Ensure labels are present
    return tokens

# Apply tokenization
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
import torch

def generate_riddle(prompt, max_length=150):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Detect device
    model.to(device)  # Move model to GPU if available

    # Tokenize input with padding and attention mask
    inputs = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True)

    input_ids = inputs.input_ids.to(device)  # Move to correct device
    attention_mask = inputs.attention_mask.to(device)  # Move attention mask to correct device

    output = model.generate(
        input_ids,
        attention_mask=attention_mask,  # Pass attention mask
        max_length=max_length,
        temperature=0.7,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id  # Explicitly set pad token
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)

# Generate 5 riddles
for _ in range(5):
    print(generate_riddle("Q:"))
```

```python
import gradio as gr
import torch

def generate_riddle(prompt, max_length=150):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    inputs = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True)
    input_ids = inputs.input_ids.to(device)
    attention_mask = inputs.attention_mask.to(device)

    output = model.generate(
        input_ids,
        attention_mask=attention_mask,
        max_length=max_length,
        temperature=0.7,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)

# Gradio Interface
demo = gr.Interface(
    fn=generate_riddle,
    inputs=gr.Textbox(label="Enter Prompt", placeholder="e.g., Q: What has hands but can't clap?"),
    outputs="text",
    title="Riddle Generator",
    description="Enter a prompt (e.g., Q:) to generate a riddle using the model."
)

demo.launch()

```
